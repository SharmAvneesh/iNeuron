```python
'''
#q1 : Try to print this by using for loop -- Done with function 
* 
* * 
* * * 
* * * * 
* * * * * 
* * * * * * 
* * * * * * * 
* * * * * * * * 
* * * * * * * * * 
'''

def printrighttringle():
            
    n=10
    for i in range(0,n):
        for j in range (0,i):
            print("* ", end=" ")
            i=i+1
        print("\n")
```

```python
'''
#q1 : Try to print this by using while loop -- Done with function 
* 
* * 
* * * 
* * * * 
* * * * * 
* * * * * * 
* * * * * * * 
* * * * * * * * 
* * * * * * * * * 
'''

def printrighttringleWhile():
    i=1
    n=10
    while i<n:
        j=1
        while j<i:
            print("* ", end=" ")
            j=j+1
        print("\n")
        i=i+1
```

```python
printrighttringleWhile()
```

```python
'''q20 : write a function by which you will be able to append two PDF files .  --> Done'''
def mergePdf():
    
    from PyPDF2 import PdfFileMerger
    pdfs=['sample1.pdf','sample2.pdf']
    merger=PdfFileMerger()
    for pdf in pdfs:
        merger.append(pdf)
        merger.write('Merged.pdf')
    merger.close()
    print("File merged successfully!")
```

```python
mergePdf()
```
